```python
%pylab --no-import-all
#%matplotlib notebook
#%matplotlib nbagg
import numpy
import matplotlib.pyplot as plt

import struct
import can
bus = can.interface.Bus(channel='can0', bustype='socketcan_native', bitrate=250000)
```

### Read from CAN ###

```python
buf = []
time = []
start = 0;
while 1:   
    packet = bus.recv()
    if packet.arbitration_id == 2:
        if start == 0:
            first_time = packet.timestamp;
            start = 1;
        time.append(packet.timestamp - first_time)
        print(packet)
        v_float = struct.unpack('f', packet.data[0:4])
        buf.append(v_float[0])

        buf.append(packet.data[4])
        buf.append(packet.data[5])
        buf.append(packet.data[6])
```

### speed + hall counter + cur u + cur w ###

```python
hall_counter = []
speed = []
cur_u = []
cur_w = []
i=0
for x in buf:
    if i == 0:
        speed.append(x)
    if i == 1:
        hall_counter.append(x)
    if i == 2:
        cur_u.append(x)
    if i == 3:
        cur_w.append(x)
        
    i = i+1
    if i == 4:
        i = 0; 
```

```python
pylab.figure()
pylab.plot(time, hall_counter, '-o', label = 'hall counter (step)')
pylab.plot(time, speed, label ='speed (rad/sec)')
pylab.grid()
pylab.xlabel('time (sec)')
pylab.legend()

pylab.figure()
pylab.plot(time, cur_u, label ='cur u')
pylab.plot(time, cur_w, label ='cur w')
pylab.grid()
pylab.xlabel('time (sec)')
pylab.legend()
```
