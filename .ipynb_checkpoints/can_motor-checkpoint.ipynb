```python
%pylab --no-import-all
#%matplotlib notebook
#%matplotlib nbagg
import numpy
import matplotlib.pyplot as plt

import struct
import can
bus = can.interface.Bus(channel='can0', bustype='socketcan_native', bitrate=250000)
```

### Read from CAN ###

```python
buf = []
time = []
start = 0;
while 1:   
    packet = bus.recv()
    if packet.arbitration_id == 0x11:
        if start == 0:
            first_time = packet.timestamp;
            start = 1;
        time.append(packet.timestamp - first_time)
        print(packet)
        v_float = struct.unpack('f', packet.data[0:4])
        buf.append(v_float[0])

        buf.append(packet.data[4])
        buf.append(packet.data[5])
        buf.append(packet.data[6])
        buf.append(packet.data[7])
```

### speed + hall counter + cur u + cur w ###

```python
time1 = time
f_v = []
i0 = []
i1 = []
i2 = []
i3 = []

i=0
for x in buf:
    if i == 0:
        f_v.append(x)
    if i == 1:
        i0.append(x)
    if i == 2:
        i1.append(x)
    if i == 3:
        i2.append(x)
    if i == 4:
        i3.append(x)
        
    i = i+1
    if i == 5:
        i = 0; 
```

```python
pylab.figure()
pylab.plot(time1, f_v, label = 'speed rad/s')
#pylab.plot(time1, i0, label ='w')
pylab.plot(time1, numpy.array(i3), label ='block time')
pylab.plot(time1, numpy.array(i2)/4,label ='temp')
pylab.grid()
pylab.xlabel('time (sec)')
pylab.legend()

#pylab.figure()
#pylab.plot(time1, f_v, label = 'speed rad/s')
pylab.plot(time1, numpy.array(i1) / 5, label ='pwm')
pylab.grid()
pylab.xlabel('time (sec)')
pylab.legend()

```
