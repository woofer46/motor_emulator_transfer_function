```python
%pylab --no-import-all
#%matplotlib notebook
#%matplotlib nbagg
import numpy
import matplotlib.pyplot as plt


import struct
import can
bus = can.interface.Bus(channel='can0', bustype='socketcan_native', bitrate=250000)
```

### Read from CAN ###

```python
motor_1_tx = 0x15
motor_1_rx0 = 0x16
motor_1_rx1 = 0x17

buf_tx0 = []
buf_tx1 = []

buf_rx0 = []
buf_rx1 = []
time_0 = []
time_set = []
start = 0;

div = 0

time_1 = []


while 1:   
    packet = bus.recv()
    if packet.arbitration_id == motor_1_rx0:
        if start == 0:
            first_time = packet.timestamp;
            start = 1;
        time_0.append(packet.timestamp - first_time)
        if div % 10 == 0:
            print(packet)
        buf_rx0.append(packet.data[0:8])
    
    if packet.arbitration_id == motor_1_rx1:
        if start == 0:
            first_time = packet.timestamp;
            start = 1;
        time_1.append(packet.timestamp - first_time)
        print(packet)
        buf_rx1.append(packet.data[0:8])
    div += 1
    
    if packet.arbitration_id == motor_1_tx:
        if start == 0:
            first_time = packet.timestamp;
            start = 1;
        time_set.append(packet.timestamp - first_time)
        buf_tx0.append(packet.data[0:8])
    
    if packet.arbitration_id == 0x01:
        if start == 0:
            first_time = packet.timestamp;
            start = 1;
        time_set.append(packet.timestamp - first_time)
        buf_tx1.append(packet.data[0:8])
```

### speed + hall counter + cur u + cur w ###

```python
time_bk_0 = time_0
speed = []
current_u = []
current_w = []
pwm = []
for x in buf_rx0:
    spd = struct.unpack('f', x[0:4])
    speed.append(spd[0])
    current_u.append(x[4])
    current_w.append(x[5])
    pwm.append((x[6] << 8) | x[7])
```

```python
time_bk_1 = time_1
temperature = []
blocked_counter = []
state = []
update_omega = []
for x in buf_rx1:
    temperature.append(x[0])
    blocked_counter.append(x[1])
    state.append(x[2])
    update_omega.append(x[3])
```

```python
time_set_0 = time_set
omega_target = []
for x in buf_tx0:
    omega_target.append(struct.unpack('f', x[0:4])[0])
```

```python
time_set_1 = time_set
pwm_target = []
for x in buf_tx1:
    pwm_x = x[2] << 8 | x[3]
    if x[1] == 2:
        pwm_x = pwm_x * -1
    pwm_target.append(pwm_x)
```

```python
pylab.figure()
pylab.plot(time_bk_0, numpy.array(speed) * 100, label = 'omega from driver rad/s')
pylab.plot(time_bk_0, numpy.array(current_u), label ='cur V')
pylab.plot(time_bk_0, numpy.array(current_w),label ='cur W')
pylab.plot(time_bk_0, numpy.array(pwm), label ='pwm from driver')

#pylab.plot(time_set_0, omega_target, label = 'target rad/s')
pylab.plot(time_set_1, pwm_target, label = 'pwm target')

pylab.grid()
pylab.xlabel('time (sec)')
pylab.legend()



pylab.figure()
pylab.plot(time_bk_1, numpy.array(temperature) / 2.3, label = 'temp')
pylab.plot(time_bk_1, numpy.array(blocked_counter), label ='block counter')
pylab.plot(time_bk_1, numpy.array(state),label = 'state')
pylab.plot(time_bk_1, numpy.array(update_omega),label = 'update flag')
pylab.grid()
pylab.xlabel('time (sec)')
pylab.legend()
```
